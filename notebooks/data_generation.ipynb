```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500

df = pd.DataFrame({
    "OrderID": range(1, n+1),
    "Date": pd.date_range(start="2022-01-01", periods=n, freq="D"),
    "Region": np.random.choice(["North", "South", "East", "West"], n),
    "Product": np.random.choice(["Laptop", "Phone", "Tablet", "Accessories"], n),
    "Sales": np.random.randint(100, 2000, n),
    "Quantity": np.random.randint(1, 5, n)
})

df.head()
```

```python
df.to_csv('../data/sales_data.csv', index=False)
```
